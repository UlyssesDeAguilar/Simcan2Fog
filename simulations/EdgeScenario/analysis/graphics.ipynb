```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def parse_if_number(s):
    try: return float(s)
    except: return True if s=="true" else False if s=="false" else s if s else None

def parse_ndarray(s):
    return np.fromstring(s, sep=' ') if s else None
```

# Caso servicio en DC

```python
# Cargamos los datos
targetDc = pd.read_csv("targetDc.csv",converters = {
    'attrvalue': parse_if_number,
    'binedges': parse_ndarray,
    'binvalues': parse_ndarray,
    'vectime': parse_ndarray,
    'vecvalue': parse_ndarray})

# Filtramos end-to-end delay +  eliminamos los NaN
vec = targetDc[targetDc.name == 'endToEndDelay:vector']
vec = vec[vec.type == "vector"]

# Empezamos el proceso de merging del conjunto de datos
row = vec.iloc[0]
mean_delay_dc = pd.DataFrame(index=row.vectime, data={'delay' : row.vecvalue*1000})

for i in range(1,len(vec)):
    row = vec.iloc[i]
    row = pd.DataFrame(index=row.vectime, data={'delay' : row.vecvalue*1000})
    mean_delay_dc = pd.concat((mean_delay_dc, row))


mean_delay_dc = mean_delay_dc.groupby(level=0).mean()
```

```python
plt.title("Latencia promedio cliente-servicio")
plt.ylabel("latencia [ms]")
plt.xlabel("tiempo simulado [s]")
plt.plot(mean_delay_dc)
plt.show()
```

# Caso servicio en Fog

```python
# Cargamos los datos
targetFog = pd.read_csv("targetFog.csv",converters = {
    'attrvalue': parse_if_number,
    'binedges': parse_ndarray,
    'binvalues': parse_ndarray,
    'vectime': parse_ndarray,
    'vecvalue': parse_ndarray})

# Filtramos end-to-end delay +  eliminamos los NaN
vec = targetFog[targetFog.name == 'endToEndDelay:vector']
vec = vec[vec.type == "vector"]

# Empezamos el proceso de merging del conjunto de datos
row = vec.iloc[0]
mean_delay_fog = pd.DataFrame(index=row.vectime, data={'delay' : row.vecvalue*1000})

for i in range(1,len(vec)):
    row = vec.iloc[i]
    row = pd.DataFrame(index=row.vectime, data={'delay' : row.vecvalue*1000})
    mean_delay_fog = pd.concat((mean_delay_fog, row), axis=1)

# Aplicamos la media
mean_delay_fog = mean_delay_fog.mean(axis=1)
```

```python
plt.title("Latencia promedio cliente-servicio")
plt.ylabel("latencia [ms]")
plt.xlabel("tiempo simulado [s]")
plt.plot(mean_delay_fog.index, mean_delay_fog, drawstyle='steps-post')
plt.show()
```

# Comparativa Dc vs Fog

```python
plt.title("Latencia promedio cliente-servicio")
plt.ylabel("latencia [ms]")
plt.xlabel("tiempo simulado [s]")
plt.plot(mean_delay_dc, drawstyle='steps-post', label="Servicio DC")
plt.plot(mean_delay_fog, drawstyle='steps-post', label= "Servicio Fog")
plt.legend(loc="upper left")
plt.show()
```
